```python
# os.mkdir('SF_features')
for file_name in name_list:
    data = pd.read_csv('features/'+file_name)
    c_len = len(data.columns)
    data_L = data.iloc[:, :int(c_len/3)]
    data_M = data.iloc[:, int(c_len/3):2*int(c_len/3)]
    data_H = data.iloc[:, 2*int(c_len/3):]

    mean_L = data_L.mean(axis = 1)
    mean_M = data_M.mean(axis = 1)
    mean_H = data_H.mean(axis = 1)
    result = pd.DataFrame(list(zip(mean_L,mean_M,mean_H)))
    sf = 'sf_'
    str_list = list(file_name)
    str_list.insert(20, sf)
    file_name_sf = ''.join(str_list)
    min_L = data_L.min(axis = 1)
    min_M = data_H.min(axis = 1)
    min_H = data_M.min(axis = 1)

    max_L = data_L.max(axis = 1)
    max_M = data_H.max(axis = 1)
    max_H = data_M.max(axis = 1)

    std_L = data_L.std(axis = 1)
    std_M = data_H.std(axis = 1)
    std_H = data_M.std(axis = 1)

    ran_L = max_L-min_L
    ran_M = max_M-min_M
    ran_H = max_H-min_H
    
    result_2 = pd.DataFrame(list(zip(min_L,max_L,mean_L,std_L,ran_L,min_M,max_M,mean_M,std_M,ran_M,min_H,max_H,mean_H,std_H,ran_H)))
    print(result_2)
    result_2.to_csv('SF_features'+'/'+file_name_sf)
```
